```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
from collections import Counter
```

```python
sd_data = pd.read_csv("Original_Dataset.csv")
```

```python
sd_data.head()
```

```python
sd_data.shape
```

```python
sc = np.array(sd_data.columns)
sc = sc[1:]
```

```python
sym = np.array(sd_data.iloc[:, 1:].values.flatten())
sym = list(set(sym))
print(sym)
```

```python
for s in sym:
    sd_data[s] = sd_data.iloc[:, 1:].apply(lambda x: int(s in x.values), axis=1)

sd_data_1 = sd_data.drop(columns=sc)
sd_data_1.head()
```

```python
sd_data_1 = sd_data_1.loc[:, sd_data_1.columns.notna()]
sd_data_1.head()
```

```python
#label_encoding the disease attribute
var_mod = ['Disease']
le = LabelEncoder()
for i in var_mod:
    sd_data_1[i] = le.fit_transform(sd_data_1[i])
```

```python
X = sd_data_1.drop(columns="Disease")
y = sd_data_1["Disease"]
```

```python
#SVM
model = svm.SVC(probability=True)
model.fit(X, y)
pred = model.predict(X)
accuracy = metrics.accuracy_score(pred, y)
joblib.dump(model, "SVC_sym.pkl")
print("Support Machine Vector",'Accuracy : %s' % '{0:.3%}'.format(accuracy))
```

```python
#Random Forest
model1 = RandomForestClassifier()
model1.fit(X, y)
pred = model.predict(X)
accuracy = metrics.accuracy_score(pred, y)
joblib.dump(model1, "Random_Forest_sym.pkl")
print("Random Forest",'Accuracy : %s' % '{0:.3%}'.format(accuracy))
```

```python
#KNN
model2 = KNeighborsClassifier()
model2.fit(X, y)
pred = model.predict(X)
accuracy = metrics.accuracy_score(pred, y)
joblib.dump(model2, "KNN_sym.pkl")
print("KNN",'Accuracy : %s' % '{0:.3%}'.format(accuracy))
```

```python
doc_data = pd.read_csv("Doctor_Versus_Disease.csv",encoding='latin1', names=['Disease','Specialist'])
```

```python
dis_data = pd.read_csv("Disease_Description.csv")
```

```python
test_col = []
for col in sd_data_1.columns:
    if col != 'Disease':
        test_col.append(col)


test_data = {}
symptoms = []
predicted = []
def test_input():
    symptoms.clear()
    predicted.clear()
    num_inputs = int(input("Enter the number of symptoms you have: "))
    for i in range(num_inputs):
        user_input = input("Enter Symptoms #{}: ".format(i+1))
        symptoms.append(user_input)
    print("Symptoms you have:", symptoms)
    for column in test_col:
        test_data[column] = 1 if column in symptoms else 0
        test_df = pd.DataFrame(test_data, index=[0])
    print("Predicting Disease based on 3 ML algorithms...")
    
    model = joblib.load('SVC_sym.pkl')
    model1 = joblib.load('Random_Forest_sym.pkl')
    model2 = joblib.load('KNN_sym.pkl')

    predict_disease = model.predict(test_df)
    predict_disease = le.inverse_transform(predict_disease)
    predicted.extend(predict_disease)
    predict_disease = model1.predict(test_df)
    predict_disease = le.inverse_transform(predict_disease)
    predicted.extend(predict_disease)
    predict_disease = model2.predict(test_df)
    predict_disease = le.inverse_transform(predict_disease)
    predicted.extend(predict_disease)
    disease_counts = Counter(predicted)

    percentage_per_disease = {disease: (count / 3) * 100 for disease, count in disease_counts.items()}
    result_df = pd.DataFrame({"Disease": list(percentage_per_disease.keys()),
                               "Chances": list(percentage_per_disease.values())})
    result_df = result_df.merge(doc_data, on='Disease', how='left')
    result_df = result_df.merge(dis_data, on='Disease', how='left')
    return result_df
```
